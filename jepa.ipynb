```python
# Create random input data (100 samples, 10 features)
X = torch.randn(100, 5)
df = pd.DataFrame(X.numpy(),columns=['Column1', 'Column2', 'Column3', 'Column4', 'Column5'])
df['Column6'] = np.random.choice([0,1], size = df.shape[0])
Y = df['Column6']
X =df.drop(columns=['Column6'])
print(X)
print(Y)

```

```python
X =df.drop(columns=['Column6']).to_numpy()
pos = np.where(Y == 1)[0]
neg = np.where(Y == 0)[0]
plt.scatter(X[pos, 0], X[pos, 1], marker='o', c='b', label='Class 1')
plt.scatter(X[neg, 0], X[neg, 1], marker='x', c='r', label='Class 0')
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.legend()
plt.show()
```
